```python
import csv

# We define a function to read the data needed to plot the charts
def readCSV(name):
    with open('data/charts/' + name) as f:
        reader = csv.reader(f)
        data = [r for r in reader]
        data.pop(0) # remove header
    return data
```

```python
import numpy as np
from numpy.polynomial.polynomial import polyfit
from sklearn.metrics import r2_score

def plotData(data, title, xlabel, ylabel) :
    import matplotlib.pyplot as plt
    
    x = np.array([ float(d[1]) for d in data ])
    y = np.array([ float(d[2]) for d in data ])

    b, m = polyfit(x, y, 1)

    plt.scatter(x,y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(x, b + m * x, '-', color='black')

    print('R squared is equal to', r2_score(y, b + m * x))
    
```

```python
weekly_displaced = readCSV("weekly_displaced.csv")

plotData(weekly_displaced, title = "One week displacement data aggregation",
         xlabel = "Sentiment Analysis Weights (week n+1)", ylabel = "Change in Ibex35 indicator (week n)")
```

```python
elmundo_weekly = readCSV("elmundo_weekly.csv")

plotData(elmundo_weekly, title = "Same week data aggregation for the newspaper El Mundo",
         xlabel = "Sentiment Analysis Weights", ylabel = "Change in Ibex35 indicator")
```

```python
elmundo_weekly_displaced = readCSV("elmundo_weekly_displaced.csv")

plotData(elmundo_weekly_displaced, title = "One week displacement data aggregation for the newspaper El Mundo",
         xlabel = "Sentiment Analysis Weights (week n)", ylabel = "Change in Ibex35 indicator (week n+1)")
```
