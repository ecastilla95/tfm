```python
# We create the Sentiment Classifier for the sentiment analysis
clf = SentimentClassifier()
```

```python
# Do we want to keep this?

def getWeights(list):
    result = [ clf.predict(e) for e in list if (clf.predict(e) >= 0.1) or (clf.predict(e) >= 0.9)]
    return result

def getAverageWeight(list):
    weighted = getWeights(list)
    length = len(weighted)
    if(length > 0):
        result = sum(weighted)/len(weighted)
        return result
    else:
        return None
```

```python
import os
basedir = "/home/osboxes/IdeaProjects/tfm/src/main/data/news/historical/elpais/"
```

```python
# We do this for all files
results = []

ls = os.listdir(basedir)
for i in ls:
        content = open(basedir+i, "r").read().splitlines()
        w = getAverageWeight(content)
        if (w != None):
            results = results + [(i, w)]
            print(i, w)
```

```python
f= open("historical.txt","w+")
for i in results:
    data = str(i[0])+"\t"+str(i[1])
    f.write("%s\n" % data)
```
