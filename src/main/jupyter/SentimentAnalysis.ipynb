```python
# We can see that there are a lot of badly classified extreme values, we can remove them

def getWeights(list):
    result = [ clf.predict(e) for e in list if (clf.predict(e) >= 0.1) or (clf.predict(e) >= 0.9)]
    return result

def getAverageWeight(list):
    weighted = getWeights(list)
    result = sum(weighted)/len(weighted)
    return result

print(getAverageWeight(x))
```

```python
import os
basedir = "/home/bigdata/IdeaProjects/tfm/src/main/data/preprocessed.news/"
```

```python
# We do this for all files
results = []

ls = os.listdir(basedir)
for i in ls:
    subls = os.listdir(basedir+i)
    for j in subls:
        case = i+"/"+j
        content = open(basedir+case, "r").read().splitlines()
        w = getAverageWeight(content)
        results = results + [(case, w)]
        print(case, w)
        
print(results)
```

```python
# Because the computation takes so long, we split it in smaller parts
elmundoList = []
elmundo = "elmundo/"
elmundols = os.listdir(basedir+elmundo)
for j in elmundols:
        case = elmundo+j
        content = open(basedir+case, "r").read().splitlines()
        w = getAverageWeight(content)
        elmundoList = elmundoList + [(case, w)]
        print(case, w)
        
print(elmundoList)
```
